# Analyse network properties for test 14

```python
import networkx as nx
import os
import operator
import re
import csv
import collections
import community
import sys

# path to where the graphs are stored
mainpath = "c:/Frederik/Afstuderen/Results/test14/results/"

# export csv files to-location
exportpath = "{}analysis/".format(mainpath)

graphs = []

for folder in os.listdir(mainpath):
    path = mainpath + folder + "/"
    
    for file in os.listdir(path):
        if file.endswith(".gexf"):
            match = re.search(r'Bible_([\w]+)\.gexf', file)
            
            if match is not None:
                graphname = [folder, match.group(1)]
                graph = nx.read_gexf(path+file, relabel=True)
                graphs.append([graphname, graph])
                print("Imported {}".format(file))
```

```python
#density = []
#avg_clust = []
network_properties = []
network_properties.append(['nr', 'book1', 'density1', 'avg_clust1', 'book2', 'density2', 'avg_clust2'])

graphnr = graphs[0][0][0]
network_properties.append([graphnr])

for graphitem in graphs:
    prev_graphnr = graphnr
    graphnr = graphitem[0][0]
    graphname = graphitem[0][1]
    graph = graphitem[1]
    
    if prev_graphnr != graphnr:
        network_properties.append([graphnr])
    
    # calculate density for this graph
    this_density = round(nx.density(graph),5)
    # calculate average clustering coefficient for this graph
    this_avg_clust = round(nx.average_clustering(graph),5)
    
    network_properties[-1].extend([graphname, this_density, this_avg_clust])
    
print("Calculated graph density and average clustering coefficient")

# Create density-table
with open("{}network_properties.csv".format(exportpath), mode="w", newline="\n") as table:
    writer = csv.writer(table)
    for network_properties_row in network_properties:
        writer.writerow(network_properties_row)
        
    print("Succesfully exported network_properties.csv")
```
